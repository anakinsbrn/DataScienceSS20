````python
#setup env
import time
import warnings
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import cluster, datasets, mixture
from sklearn.neighbors import kneighbors_graph
from sklearn.preprocessing import StandardScaler
````

````python
#helper calls
n_samples = 1500
random_state = 170
X, y = datasets.make_blobs(n_samples=n_samples, random_state=random_state)
transformation = [[0.6, -0.6], [-0.4, 0.8]]
X_aniso = np.dot(X, transformation)

#generating Data Sets A-F
A = datasets.make_circles(n_samples=n_samples, factor=.5,noise=.05)[0]
B = datasets.make_moons(n_samples=n_samples, noise=.05)[0]
C = datasets.make_blobs(n_samples=n_samples, random_state=8)[0]
D = np.random.rand(n_samples, 2)
E = (X_aniso, y)[0]
F = datasets.make_blobs(n_samples=n_samples,
                             cluster_std=[1.0, 2.5, 0.5],
                             random_state=random_state)[0]
````

## Exercise 1
Plot all raw data sets A-F in one figure.

Hints: use ```scatter``` plots and ```subfigures```.

````python
plt.figure(figsize=(8, 8))
plt.subplots_adjust(bottom=.05, top=.9, left=.05, right=.95)
plt.subplot(321)
plt.scatter(A[:,0], A[:,1])
plt.subplot(322)
plt.scatter(B[:,0], B[:,1])
plt.subplot(323)
plt.scatter(C[:,0], C[:,1])
plt.subplot(324)
plt.scatter(D[:,0], D[:,1])
plt.subplot(325)
plt.scatter(E[:,0], E[:,1])
plt.subplot(326)
plt.scatter(F[:,0], F[:,1])
````

## Exercise 2
Perform ```K-Means``` clustering on all data sets: https://scikit-learn.org/stable/modules/clustering.html#k-means
* 2.1 plot all results
* 2.2 manually try to find the best $k$ for each data set (by visual evaluation)
* 2.3 Use the ```K-Means++``` initialization 

````python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

kmeansA = KMeans(n_clusters=2, init='k-means++', random_state=0).fit(A)
kmeansB = KMeans(n_clusters=3, init='k-means++', random_state=0).fit(B)
kmeansC = KMeans(n_clusters=3, init='k-means++', random_state=0).fit(C)
kmeansD = KMeans(n_clusters=2, init='k-means++', random_state=0).fit(D)
kmeansE = KMeans(n_clusters=3, init='k-means++', random_state=0).fit(E)
kmeansF = KMeans(n_clusters=3, init='k-means++', random_state=0).fit(F)

plt.figure(figsize=(8, 8))
plt.subplots_adjust(bottom=.05, top=.9, left=.05, right=.95)
plt.subplot(321)
plt.scatter(A[:,0], A[:,1],c=kmeansA.labels_)
plt.subplot(322)
plt.scatter(B[:,0], B[:,1],c=kmeansB.labels_)
plt.subplot(323)
plt.scatter(C[:,0], C[:,1],c=kmeansC.labels_)
plt.subplot(324)
plt.scatter(D[:,0], D[:,1],c=kmeansD.labels_)
plt.subplot(325)
plt.scatter(E[:,0], E[:,1],c=kmeansE.labels_)
plt.subplot(326)
plt.scatter(F[:,0], F[:,1],c=kmeansF.labels_)

````

## Exercise 3
Performs ```DBSCAN``` clustering on all data sets and plot all results: https://scikit-learn.org/stable/modules/clustering.html#dbscan

````python
from sklearn.cluster import DBSCAN

db_A = DBSCAN(eps=0.6, min_samples=2).fit(A)
db_B = DBSCAN(eps=0.2, min_samples=2).fit(B)
db_C = DBSCAN(eps=1.5, min_samples=2).fit(C)
db_D = DBSCAN(eps=0.1, min_samples=2).fit(D)
db_E = DBSCAN(eps=0.3, min_samples=4).fit(E)
db_F = DBSCAN(eps=0.5, min_samples=5).fit(F)


plt.figure(figsize=(8, 8))
plt.subplots_adjust(bottom=.05, top=.9, left=.05, right=.95)
plt.subplot(321)
plt.scatter(A[:,0], A[:,1],c=db_A.labels_)
plt.subplot(322)
plt.scatter(B[:,0], B[:,1],c=db_B.labels_)
plt.subplot(323)
plt.scatter(C[:,0], C[:,1],c=db_C.labels_)
plt.subplot(324)
plt.scatter(D[:,0], D[:,1],c=db_D.labels_)
plt.subplot(325)
plt.scatter(E[:,0], E[:,1],c=db_E.labels_)
plt.subplot(326)
plt.scatter(F[:,0], F[:,1],c=db_F.labels_)


````
